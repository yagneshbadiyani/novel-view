```python
import numpy as np
import scipy.misc as spm
import tensorflow as tf
import matplotlib.pyplot as plt
import os
```

```python
basedir='/datasets/yagneshb/kitti/sequences/'
scenes=os.listdir(basedir)
```

```python
for scene in scenes:
    os.chdir(basedir+scene+'/image_2')
    images=os.listdir(os.curdir)
    for image in images:
        I=plt.imread(image)
        I=spm.imresize(I,[128,416])
        spm.imsave('../'+image,I)
```
